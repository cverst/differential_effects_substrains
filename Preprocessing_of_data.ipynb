Import packages.

```python
import pandas as pd
import os
from utils.preprocessing import abr_prep, dp_prep
```

# ABR

## Load data
Read raw data from csv.

```python
data_path_caj_abr = os.path.join(os.getcwd(), "data", "raw", "ABR_CaJ_raw.csv")
data_path_jrj_abr = os.path.join(os.getcwd(), "data", "raw", "ABR_JRj_raw.csv")
data_path_sca_abr = os.path.join(os.getcwd(), "data", "raw", "ABR_Sca_raw.csv")
```

```python
df_caj_abr = pd.read_csv(data_path_caj_abr)
df_jrj_abr = pd.read_csv(data_path_jrj_abr)
df_sca_abr = pd.read_csv(data_path_sca_abr)
```

## Preprocessing

```python
df_caj_abr_prep = abr_prep(df_caj_abr)
```

```python
df_jrj_abr_prep = abr_prep(df_jrj_abr)
```

```python
df_sca_abr_prep = abr_prep(df_sca_abr)
```

## Save
Write data to csv.

```python
prep_data_dir = os.path.join(os.getcwd(), "data", "preprocessed")
if not os.path.exists(prep_data_dir):
    os.mkdir(prep_data_dir)
```

```python
df_caj_abr_prep.to_csv(os.path.join(prep_data_dir, "ABR_CaJ_prep.csv"), index=False)
df_jrj_abr_prep.to_csv(os.path.join(prep_data_dir, "ABR_JRj_prep.csv"), index=False)
df_sca_abr_prep.to_csv(os.path.join(prep_data_dir, "ABR_Sca_prep.csv"), index=False)
```
